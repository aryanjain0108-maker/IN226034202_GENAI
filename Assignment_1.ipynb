```python
#Task 2: Build Advanced Preprocessing Function
import re
import string
import nltk
from nltk.tokenize import word_tokenize
def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    tokens = word_tokenize(text)
    filtered_tokens = []
    for word in tokens:
        if len(word) > 2 or word in ["no", "not"]:
            filtered_tokens.append(word)
    cleaned_sentence = " ".join(filtered_tokens)
    return filtered_tokens, cleaned_sentence
```

```python
#Task 3: Stress Testing
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
results = []
for text in test_sentences:
    tokens, sentence = preprocess_text(text)
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_len = sum(len(t) for t in tokens) / total_tokens if total_tokens > 0 else 0
    results.append({
        "original": text,
        "tokens": tokens,
        "cleaned": sentence,
        "total_tokens": total_tokens,
        "unique_tokens": unique_tokens,
        "avg_length": avg_len
    })
for r in results:
    print("\nOriginal:", r["original"])
    print("Tokens:", r["tokens"])
    print("Cleaned:", r["cleaned"])
    print("Total Tokens:", r["total_tokens"])
    print("Unique Tokens:", r["unique_tokens"])
    print("Avg Token Length:", round(r["avg_length"], 2))
```

```python
#Task 4: Token Analytics
noisiest = min(results, key=lambda x: x["total_tokens"])
meaningful = max(results, key=lambda x: x["unique_tokens"])
print("Most Noisy Sentence:", noisiest["original"])
print("Most Meaningful Sentence:", meaningful["original"])
```

```python
#Task 5: Frequency Analysis
from collections import Counter
all_tokens = []
for r in results:
    all_tokens.extend(r["tokens"])
counter = Counter(all_tokens)
print("Top 10 Most Frequent Words:")
print(counter.most_common(10))
print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

```python
#Task 6: Build Full Pipeline
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []
    for text in text_list:
        tokens, sentence = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(sentence)
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
output = full_pipeline(test_sentences)
print(output)
```

```python
#Task 7: Error Handling
edge_cases = [
    "",
    "😂😂😂😂",
    "1234567890"
]
for text in edge_cases:
    tokens, sentence = preprocess_text(text)
    print("\nInput:", text)
    print("Tokens:", tokens)
    print("Cleaned:", sentence)
```
